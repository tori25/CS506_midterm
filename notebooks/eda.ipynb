```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

trainingSet = pd.read_csv("../data/train.csv")
testingSet = pd.read_csv("../data/test.csv")


numeric_data = trainingSet.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()

numeric_data['Score'] = trainingSet['Score']

corr_with_target = numeric_data.corr()['Score'].sort_values(ascending=False)
print(corr_with_target)

top_features = corr_with_target.index[1:6]

for col in top_features:
    sns.boxplot(x=trainingSet['Score'], y=trainingSet[col])
    plt.title(f"{col} vs Score")
    plt.show()

print(trainingSet['Score'].value_counts(normalize=True))

print(trainingSet.shape)
print(trainingSet.columns)
print(trainingSet.dtypes)

print(trainingSet.isnull().sum().sort_values(ascending=False).head(10))

print(trainingSet.describe())

print("train.csv shape is ", trainingSet.shape)
print("test.csv shape is ", testingSet.shape)

print()

print(trainingSet.head())
print()
print(testingSet.head())

print()

print(trainingSet.describe())

trainingSet['Score'].value_counts().plot(kind='bar', legend=True, alpha=.5)
plt.title("Count of Scores")
plt.show()
```
